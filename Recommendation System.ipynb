```python
# Drop unnamed column
ratings.drop('timestamp', axis=1, inplace=True)
```

```python
df = ratings
```

```python
print(df)
```

```python
#getting user id and appending to list
#getting movie id and appending to list
#getting rating value and appending to list
user_data=[]
movie_data=[]
ratings_data=[]
for i in range(df.shape[0]):
    user=(df['userId'].iloc[i])                
    movie=(df['movieId'].iloc[i])               
    rat=(df['rating'].iloc[i])                  
    user_data.append(user)
    movie_data.append(movie)
    ratings_data.append(rat)
```

```python
from scipy.sparse import csr_matrix
adj_matrix = csr_matrix((ratings_data, (user_data, movie_data)))
adj_matrix.shape
```

```python
print(adj_matrix)
```

```python
from sklearn.utils.extmath import randomized_svd
import numpy as np 

U, S, VT = randomized_svd(adj_matrix, n_components=5,n_iter=5, random_state=None)
print(U.shape)
print(S.shape)
print(VT.T.shape)
```
